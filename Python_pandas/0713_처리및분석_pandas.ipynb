## 실습문제 2

```python
data = {'weight':[91,85,100,75,43,38],
       'height':[180,173, 190, 170, 163, 152],
       'gender':['male','female','male','female','female','male']}

df = DataFrame(data)
df
```

```python
#2 weight가 80 이상
df_filtered = df[df.weight >=80]
df_filtered
```

```python
#3 weigth 기준 오름차순으로 정렬, 몸무게가 적은 순으로 정렬
df_sorted = df_filtered.sort_values('weight', ascending=True)
print(df_sorted)

df_sorted = df_filtered.sort_values('height', ascending=False)
print(df_sorted)
```

```python
#4 height 기준으로 내림차순으로 정렬, 키가 큰 순으로 정렬
#5 성별에 따른 평균
df_grouped = df.groupby('gender') #'gender' 값으로 데이터들을 그룹화
df_grouped_mean = df_grouped.mean() #grouped.mean()그룹들의 평균
print(df_grouped_mean)

for key, group in df_grouped:
    print(key) # 그룹을 묶을 때 쓴 키를 출력
    print(group) # 키로 묶인 그룹을 출력
```
