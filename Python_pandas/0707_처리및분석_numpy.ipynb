### lab : MSE 오차 계산하기

```python
#회귀 문제나 분류 문제에서 실제 출력과 우리가 원하는 출력간의 오차를 계산하기 위해 MSE(Mean Square Error)를 계산

ypredi = np.array([1,0,0,0,0])
yi = np.array([0,1,0,0,0])
n= 5

MSE = (1/n) * np.sum(np.square(ypredi-yi))

MSE
```
